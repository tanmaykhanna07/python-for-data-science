```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(42)
num_rows = 10000

# Generate 10,000 rows of banking transactions
df_bank = pd.DataFrame({
    "Transaction_ID": range(100000, 100000 + num_rows),
    "Amount": np.round(np.random.exponential(scale=50, size=num_rows) + 5, 2),
    "Location": np.random.choice(["Mumbai", "Delhi", "Bangalore", "Chennai", "Pune"], size=num_rows),
    "Account_Age_Days": np.random.randint(1, 3650, size=num_rows),
    "Device": np.random.choice(["Mobile", "Desktop", "Tablet"], size=num_rows, p=[0.7, 0.25, 0.05])
})

# Inject a "Fraud" flag: True if the amount is high AND the account is brand new
df_bank["Is_Fraud"] = ((df_bank["Amount"] > 150) & (df_bank["Account_Age_Days"] < 30))

print("Total Rows:", len(df_bank))
df_bank.head()
```

```python
frauds = df_bank[df_bank['Is_Fraud'] == True]
print("Total Frauds: ", len(frauds))
```

```python
mean_amount_by_city = df_bank.groupby('Location')['Amount'].mean()
mean_amount_by_city
```

```python
high_desktop_trans = df_bank[(df_bank['Device'] == "Desktop") & (df_bank['Amount'] > 300) ]
high_desktop_trans
```
