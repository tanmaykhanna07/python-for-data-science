```python
import pandas as pd
```

```python
server_logs = {
    "Timestamp": ["10:00", "10:05", "10:10", "10:15", "10:20"],
    "Endpoint": ["/login", "/get_habits", "/login", "/update_habit", "/get_habits"],
    "Status_Code": [200, 200, 401, 500, 200],
    "Response_Time_ms": [45, 120, 30, 850, 110]
}

df = pd.DataFrame(server_logs)

df
```

```python
response_times = df['Response_Time_ms']
print(response_times)
```

```python
is_crash = df['Status_Code'] == 500
crashed_requests = df[is_crash]
print(crashed_requests)
```

```python
#industry standard of [4]

crashed_requests = df[df['Status_Code'] == 500]
print(crashed_requests)
```

```python
endpoint_performance = df.groupby('Endpoint')['Response_Time_ms'].mean()

print("Average Response Time per Endpoint (ms):")
print(endpoint_performance)
```

```python
#Feature Engineering - create new data out of old data
df['Is_Slow'] = df['Response_Time_ms'] > 100

sorted_requests = df.sort_values(by='Response_Time_ms', ascending= False)

sorted_requests
```

```python
checkout_data = {
    "User_ID": [101, 102, 101, 103, 102, 101],
    "Item": ["Mechanical Keyboard", "Monitor", "Mouse", "Laptop", "Headset", "Mousepad"],
    "Price": [150, 300, 50, 1200, 80, 20],
    "Status": ["Success", "Failed", "Success", "Success", "Success", "Failed"]
}

df_sales = pd.DataFrame(checkout_data)
df_sales
```

```python
total_revenue_by_user = df_sales.groupby('User_ID')['Price'].sum()
total_revenue_by_user
```

```python
successfull_purchases = df_sales[df_sales['Status'] == "Success"]
total = successfull_purchases.groupby('User_ID')['Price'].sum()
total
```

```python
failed_purchases = df_sales[df_sales['Status'] == "Failed"]
failed_purchases
```

```python
df_sales['Is_Expensive'] = df_sales['Price'] >= 100
sorted_df = df_sales.sort_values(by = "Price" , ascending = False)
sorted_df
```
